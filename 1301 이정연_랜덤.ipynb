```python
import random as r
a=r.randint(1,7)
b=r.randint(1,7)
list_c=[a,b]
u=input('숫자두개 입력:')
u=list(map(int,u.split(' ')))
print('당신의 선택',u)
print('컴퓨터의 선택',list_c)
d=0
for i in u:
    if i in list_c:
        d+=1
if d==2:
    print('1등')
elif d==1:
    print('2등')
else:
    print('3등')
```

```python

a=input('생년월일과 성별정보를 입력하세요:')
b=list(map(int,a.split()))
print(b)
year = b[0] // 10000 
if b[1]==1:
    print('1900 년대 남자')
    year+=1900
    print('현재나이는',2027-year)
elif b[1]==2:
    year+=1900
    print('1900 년대 여자')
    print('현재나이는',2027-year)
elif b[1]==3:
    year+=2000
    print('2000년대 남자')
    print('현재나이는',2027-year)

elif b[1]==4:
    year+=2000
    print('2000년대 여자')
    print('현재나이는',2027-year)
```

```python
from datetime import datetime
ymd,s = input().split(',')
yy = int(ymd[0:2])
mm = int(ymd[2:4])
dd = int(ymd[4:6])
s = int(s)
if s == 1 :
    year= 1900+yy
    sex = '남성'
elif s == 2 :
    year= 1900+yy
    sex = '여성'
elif s == 3 :
    year = 2000+yy
    sex = '남성'
elif s == 4:
    year = 2000+yy
    sex = '여성'

today = datetime.now()
age = today.year-year

if (today.month,today.day)<(mm,dd) :
    age -= 1
print(f'현재나이 {age}살 {sex}입니다. ')
                            
```
