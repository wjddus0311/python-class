```python
s=1
hap=0
c=0
while True:
    hap+=s
    s+=1
    c+=1
    if s>10:
        break
print(hap)
print(hap/c)
```

```python
score=0
hap=0
avg=0
cnt=0
while True:
    score=input("점수를 입력하세요: ")
    if score=='q':
        print("총점: %.1f, 평균: %.1f"%(hap,avg))
        break
    else:
        hap+=float(score)
        cnt+=1
        avg=hap/cnt
```
