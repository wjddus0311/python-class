```python
#8
a=int(input('연필의 판매 개수는?'))
d=a*700
b=int(input('지우개의 판매 개수는?'))
e=b*500
c=int(input('공책의 판매 개수는?'))
f=c*2500
g=d+e+f
print('------------------------')
print('파이썬 문방구의 오늘 판매 내역')
print('------------------------')
print("1. 연필:",a,'개',d,'원')
print("2. 지우개:",b,'개',e,'원')
print("3. 공책:",c,'개',f,'원')
print('------------------------')
print("오늘 총 판매 금액:",g,'원')
print('------------------------')

```
