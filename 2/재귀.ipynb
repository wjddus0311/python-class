```python
def factorial(n):
  if n <= 1:
    return 1
  return n * factorial(n - 1)
print(factorial(5))
```

```python
def fio(n):
  if n <= 1:
    return n
  return fio(n - 1) + fio(n - 2)
print(fio(10)) 
```

```python
def hanoi(n, start, end):
  if n == 1:
    print(start, '->', end)
    return
  hanoi(n - 1, start, 6 - start - end)
  print(start, '->', end)
  hanoi(n - 1, 6 - start - end, end)
hanoi(3, 1, 3)
```

```python
def hanoi(n,start,end,sub):
  if n == 1:
    print(start, '->', end)
    return
  hanoi(n - 1, start, sub, end)
  print(start, '->', end)
  hanoi(n - 1, sub, end, start)
n=int(input())
hanoi(n, 1, 3, 2)
```

```python
def palin(s):
  s=s.lower()
  if s == s[::-1]:
    print('true')
  else:
    print('false')
s=input()
palin(s)
```

```python
def palin(s):
  if len(s) <= 1:
    return True
  if s[0].lower() != s[-1].lower():
    return False
  return palin(s[1:-1])
s=input()
if palin(s):
  print('true')
else:
  print('false')
```
