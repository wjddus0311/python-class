```python
#3
c=0
t=0
while True:
    s=int(input())
    if s==-1:
        break
    if 0<=s<=100:
        t+=s
        c+=1
    else:
        print('잘못된 결과')
if c > 0:
    a = t / c
    print("평균:", a)
else:
    print("입력된 점수가 없습니다")

```

```python
has_ex=False
has_n=False
while True:
    p=input('비밀번호')
    for i in p:
        if i=='!':
            has_ex=True
        if i.isdigit():
            has_n=True
    if len(p)>=8 and has_ex and has_n:
        print('사용가능')
        break
    else:
        print('잘못 입력')
```
