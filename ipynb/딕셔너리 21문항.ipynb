```python
n = int(input())
scores = []
for i in range(n): #range(5) = 0,1,2,3,4
    s = int(input())
    scores.append(s)
high_s = 0
total = 0
low_s = 100
ave = 0
for i in scores:
    total += i
    if i > high_s:
        high_s = i
    if i < low_s:
        low_s = i
ave = int(total/n)
print(f'총점: {total}')
print(f'평균: {ave}')
print(f'최고: {high_s}')
print(f'최저: {low_s}')
```

```python
w=input()
n={'alpha':0,'digit':0,'special':0}
for i in w:
  key=i.lower()
  if key.isalpha():
    n['alpha']+=1
  elif key.isdigit():
    n['digit']+=1
  else:
    n['special']+=1
print(n)
```
