```python
#클래스 정의
class Monster:
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def say(self):
    print(f'나는 {self.name}, {self.age}살 이다')
#객체 생성
shark=Monster('상어',3)
wolf=Monster('늑대',5)

#메서드 호출
shark.say()
wolf.say()
```

```python
class Student:
  def __init__(self,name,age,score):
    self.name=name
    self.age=age
    self.score=score
  def is_pass(self):
    return self.score>=60
stu1=Student('민수',17,90)
stu2=Student('영희',17,65)
stu3=Student('james',17,50)

print(stu1.is_pass())
print(stu2.is_pass())
print(stu3.is_pass())
```

```python
class Dog:
  def __init__(self,name,age,type):
    self.name=name
    self.age=age
    self.type=type
  def bark(self):
    return print(f'{self.type} 멍멍')
  def sit(self):
    return print(f'{self.name} 앉아!')

dog1=Dog('당근',5,'진도')
dog2=Dog('망고',3,'푸들')
dog1.bark()
dog2.sit()
dog1.sit()
dog2.bark()
```

```python
class Animal:
  def speak(self):
    print('동물이 소리를 냅니다.')
class Dog(Animal):
  def speak(self):
    print('멍멍')
dog=Dog()
dog.speak()
```

```python
class FishBread:
  def __init__(self,content):
    self.content=content
  def cook(self):
    print(f'붕어빵을 굽습니다')
  def eat(self):
    print(f'{self.content}붕어빵을 맛있게 먹습니다')
class IceFishBread(FishBread):
  def freeze(self):
    print('아이스크림을 넣습니다')
  def eat(self):
    print(f'{self.content}붕어빵에 아이스크림을 넣어 차갑게 먹습니다')
bread=IceFishBread('슈크림')
bread.cook() #부모기능 재사용
bread.freeze() #자식 기능 추가, 새 기능 추가
bread.eat()

fish=FishBread('딸기잼')
fish.eat()
```

```python
class BankAccount:
  def __init__(self,owner,balance): #balance=잔액
    self.owner=owner
    self.__balance=balance #private 변수로 설정
  def deposit(self,amount):
    if amount>0:
      self.__balance+=amount
      print(f'{amount}원이 입금되었습니다.')
    else:
      print('입금액은 0원보다 커야 합니다.')
  def get_balance(self):
    return self.__balance
account=BankAccount('철수',1000)
account.deposit(5000)
account.__balance=10000 #private 변수는 외부에서 직접 접근 불가
print(account.get_balance())
print(account.__dict__) #private 변수 접근 방법
```

```python
class BankAccount:
  def __init__(self,owner,balance): #balance=잔액
    self.owner=owner
    self.balance=balance
account=BankAccount('철수',1000)
account.balance=-5000
print(account.balance)
```

```python
class Student:
  def __init__(self,name,score):
    self.name=name
    self.__score=score
  def get_score(self):
    return self.__score
  def set_score(self,score):
    if 0<=score<=100:
      self.__score=score
    else:
      print('점수는 0에서 100 사이')

s1=Student('철수',80)
print(s1.get_score()) #private 변수 접근 불가
s1.set_score(95) #setter 메서드로 점수 수정
print(s1.get_score())
s1.set_score(150) #잘못된 점수 입력
print(s1.get_score()) #점수는 수정되지 않음

```
