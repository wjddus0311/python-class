```python
def triangle(width,height):
  area=(width*height)/2
  return area

width=input("삼각형의 밑변:")
width=int(width)
height=input("삼각형의 높이:")
height=int(height)
area=triangle(width,height)
print("삼각형의 넓이:",area)
```

```python
def q(a):
  t=a*3600
  return t
hour=input('변환하려는 시간(h)')
hour=int(hour)
second=q(hour)
print(hour,'시간은',second,'초입니다')
```

```python
def list_avg(a):
  s=0
  for i in a:
    s+=i
  result=s/len(a)
  print('점수의 평균',result) 
score=[90,80,70,50,60]
list_avg(score) 
```

```python
#학습지 53번 
def lst_create():
   lst=list(map(int,input().split()))
   print(lst)
   return lst
def lst_append(lst_):
    while True:
      b=(input('추가할 숫자:'))
      try:
        lst_.append(float(b))
      except:
        break
    return lst_
    
def lst_cal(lst):
  c=0
  for i in lst:
    c+=i
  avg=c/len(lst)
  return c,avg
    
def lst_print(lst,result):
  print('점수',lst)
  print('합계:',result[0])
  print('평균:',result[1])

lst=lst_create()
lst=lst_append(lst)
print(lst)
result=lst_cal(lst)
lst_print(lst,result)

```
