```python
#문제1
def lst_create():
  a=list(map(int,input().split()))
  print(a)
  return a
def lst_append(lst):
    while True:
       a=(input())
       if a=='q':
        print(lst)
        return lst
       lst.append(int(a))

def lst_cal(lst):
  total=sum(lst)
  avg=total/len(lst)
  return total,avg

def lst_print(total,avg):
  print(f'점수 : {lst}')
  print(f'합계 : {total} 평균 : {avg}')







lst=lst_create()
lst=lst_append(lst)
result=lst_cal(lst)
print(result)
lst_print(result[0],result[1])




```

```python
def sub_1(lst):
  mylst = list(range(7,11))
  print(f'sub_1의 함수리스트 : {mylst}')
def sub_2(lst):
  global mylst
  mylst = list(range(1,5))
  print(f'sub_2의 함수리스트 : {mylst}')

mylst = list(range(10,50,10))
sub_1(mylst)
sub_2(mylst)
print(f'메인함수 리스트: {mylst}')
```

```python
#클래스 문제
class FishBread:
  kind='뿡뿡이 붕어빵'
  def __init__(self,content):
    self.content=content
  def cook(self):
    print('붕어빵을 굽습니다')
    print('붕어: 으아아악!')
  def eat(self):
    print(f'{self.content}붕어빵을 맛있게 먹습니다!')

class IceFishBread(FishBread):
  def freeze(self):
    print('아이스크림을 넣습니다')
  def eat(self):
    print(f'{self.content}붕어빵에 아이스크림을 넣어 차갑게 먹습니다')
    print('붕어: 아 차가워ㅜㅜ')
    
bread=IceFishBread('슈크림')
bread.freeze()
bread.eat()

print(FishBread.kind)
fish = FishBread('딸기잼')
fish.cook()
fish.eat()



  

    

```
