```python
f=open('testx.csv','r',encoding='utf-8-sig')
rows=(f.read()).split('\n')
f.close()
print(rows)

for i in rows:
  v=i.split(',')
  # print(v)
  for j in v:
    print(j,end=' ')
  print()
```

```python
f=open('testx2.csv','r',encoding='utf-8-sig')
rows=(f.read()).splitlines()
f.close()
print(rows)

for i in rows:
  v=i.split(',')
  sum=0
  for j in v:
    sum+=int(j)
  print(sum)
```
