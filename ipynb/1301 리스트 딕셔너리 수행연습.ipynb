```python
#5
c = int(input())
stu = []
high = 0
low = 10**20
total = 0
for i in range(c):
    o = int(input())
    stu.append(o)

for i in stu:
    total += i
    if i > high:
        high = i
    if i < low:
        low = i
print('총점:', total)
print('평균:', int(total/c))
print('최고:', high)
print('최저:', low)
```

```python
#6
b={}
a=input()
for i in range(int(a)):
  a,s=input().split()
  b[a]=int(s)
  na=None
  m=-1
  for a,s in b.items():
    if s>m:
      m=s
      na=a
  print(na,m)
  
  for a in b:
    if b[a]>m:
      m=b[a]
      na=a
  print(na,m)
```

```python
a=input()
seen=set()
dup=set()
for i in a:
  if i in seen:
    dup.add(i)
  else:
    seen.add(i)
print(dup)
```

```python
words =input().split()
result={}
for w in words:
  w=w.lower()
  if w in result:
    result[w]+=1
  else:
    result[w]=1
for i in result:
  print(i,result[i])
```

```python
words =input().split()
result={}
for w in words:
  w=w.lower()
  result[w]=result.get(w,0)+1
for k,v in result.items():
  print(k,v)  
```

```python
w=input()
n={'alpha':0,'digit':0,'special':0}
for i in w:
  key=i.lower()
  if key.isalpha():
    n['alpha']+=1
  elif key.isdigit():
    n['digit']+=1
  else:
    n['special']+=1
print(n)
```
