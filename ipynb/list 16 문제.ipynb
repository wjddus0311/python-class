```python

nu = list(map(int, input("숫자들을 입력하세요: ").split()))
m = nu[0]
min = nu[0]
for n in nu:
    if n > m:
        m = n
    if n < min:
        min = n
print("최댓값:", m)
print("최솟값:", min)
```

```python

a = int(input("학생 수를 입력하세요: "))

scores = []
b = 0
for i in range(a):
    score = int(input())
    scores.append(score)
    if score >= 80:
        b += 1
print(f"평균 점수: {sum(scores) / a:.1f}")  
print(f"80점 이상 학생 수: {b}명")
```

```python
a=int(input())
b={}
t=0
for i in range(a):
  n,s=input().split()
  b[n]=s
  t+=int(s)
  avg=t//a
for n,s in b.items():
  if int(s)>=avg:
    print(n,s)
```

```python
a=input().split()
b={}
for i in a:
  if '.' in i:
    e=i.split('.')[1].lower()
    if e in b:
      b[e]+=1
    else:
      b[e]=1
for e,c in b.items():
  print(f"{e}: {c}회")
```
