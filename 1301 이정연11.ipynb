```python
list=['홍길동','일지매']
print('현재 프로그래밍 수강 신청자는',list,'입니다.')
b=input('수강신청할 이름을 입력하세요')
list.append(b)
print(b, '학생의 신청이 완료되었습니다.')
print('현재 이과목의 최종 신청자는',list,'입니다')
c=input('삭제할 학생의 이름을 입력하세요')
list.remove(c)
print( c ,'학생의 수강 철회가 완료되었습니다.')
print('현재 이 과목의 최종 신청자는',list,'입니다.')
print('현재 이과목의 최종 신청자는',list,'입니다.')
list.remove('일지매')
a=input('수정할 이름을 입력하세요:')
list.insert(0,a)
print('요청 하신대로 일지매를',a,'로 변경하였습니다.')
print('현재 이괌고의 최종 신청자는',list,'입니다.')
```
